## 1\. Pinguins

Neste exercício, vamos utilizar uma base de dados com informações sobre penguins. A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```
import numpy as np
import pandas as pd
import seaborn as sns
import re
```

```
data = sns.load_dataset('penguins')
```

```
data
```

### **1.1. Valores nulos**

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

```
# convertendo o dataset num dataframe e ferificando a % de valores nulos
df = pd.DataFrame(data)
missinss_values = df.isnull().mean()*100
print(missinss_values)
```

```
# iniciando o tratamento removedo o maior % de nulos ('sex')
remove_null = df.dropna(subset=['sex'])
df = remove_null
```

```
# com a remoção da coluna de maior%, as outras também foram tratadas
missinss_values = df.isnull().mean()*100
print(missinss_values)
```

### **1.2. Variáveis numéricas**

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

```
# visalizando as variáveis númericas
df[['bill_length_mm','bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
```

```
# criando um função para padronizar as colunas numéricas
def calcular_std(df,column):
  mean = df[column].mean()
  std = df[column].std()
  df[f'{column}_std'] = (df[column]-mean)/std
  return df
```

```
# aplicando em cada uma a funcão
lenght = calcular_std(df,'bill_length_mm')
depth = calcular_std(df,'bill_depth_mm')
flipper = calcular_std(df,'flipper_length_mm')
mass = calcular_std(df, 'body_mass_g')
```

### **1.3. Variáveis categóricas**

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

```
# visuzalizando as variáveis categóricas
df[['island', 'sex']]
```

```
# utilizando o .get_dummies para aplicar 'one hot encoding'
island = pd.get_dummies(df['island'])
island.columns = ['island' +'_'+ str(col) +'_nom' for col in island.columns]
df = pd.concat([df, island], axis=1)
df
```

```
# aplicando a mesma ténica anterior
sex = pd.get_dummies(df['sex'])
sex.columns = [str(col) +'_nom' for col in sex.columns]
df = pd.concat([df, sex], axis=1)
df
```

### **1.4. Limpeza**

Descarte as colunas originais e mantenha apenas a variável resposta e as variáveis preditivas com o sufixo *_std*", *_nom*" e "*_ord*".

```
# identificando as colunas de interesse (species + *_std + *_nom)
column_final = ['species']

for col in df:
  if re.findall(".*_std.*|.*_nom.*", col):
    column_final.append(col)

df_final = df[column_final]
```
